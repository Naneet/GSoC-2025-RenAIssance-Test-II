```python
processor = TrOCRProcessor.from_pretrained("qantev/trocr-large-spanish")
model = VisionEncoderDecoderModel.from_pretrained("qantev/trocr-large-spanish").to(device)
model.eval()
```

```python
def get_bboxes(image_name):
    name = image_name.split(".")[0]
    bboxes_file_name = "res_" + name + ".txt"
    bboxes_file_path = os.path.join(bboxes_path, bboxes_file_name)

    bboxes_raw = open(bboxes_file_path, 'r', encoding="utf-8").read()
    bboxes_strings = bboxes_raw.split("\n\n")

    if bboxes_strings[-1] =="":
        bboxes_strings.pop()

    bboxes = []
    for bboxes_string in bboxes_strings:
        bboxes_string_split = bboxes_string.split(",")
        bboxes.append((int(bboxes_string_split[0]), int(bboxes_string_split[1]), int(bboxes_string_split[4]), int(bboxes_string_split[5])))

    return bboxes


def merge_bounding_boxes(bboxes, x_thresh=150, y_thresh=10):
    if not bboxes:
        return []

    bboxes = np.array(bboxes)
    bboxes = bboxes[bboxes[:, 1].argsort()]
    merged = []
    visited = np.zeros(len(bboxes), dtype=bool)  # Track visited boxes

    def find_cluster(idx):
        cluster = [idx]
        x1, y1, x2, y2 = bboxes[idx]
        x_center, y_center = (x1 + x2) / 2, (y1 + y2) / 2

        for j in range(len(bboxes)):
            if j != idx and not visited[j]:
                x1_j, y1_j, x2_j, y2_j = bboxes[j]
                x_center_j, y_center_j = (x1_j + x2_j) / 2, (y1_j + y2_j) / 2

                dx = abs(x_center - x_center_j)
                dy = abs(y_center - y_center_j)

                if dx <= x_thresh and dy <= y_thresh:
                    visited[j] = True
                    cluster.extend(find_cluster(j))

        return cluster

    for i in range(len(bboxes)):
        if not visited[i]:
            visited[i] = True
            cluster = find_cluster(i)
            merged_x1 = np.min(bboxes[cluster, 0])
            merged_y1 = np.min(bboxes[cluster, 1])
            merged_x2 = np.max(bboxes[cluster, 2])
            merged_y2 = np.max(bboxes[cluster, 3])
            merged.append((merged_x1, merged_y1, merged_x2, merged_y2))

    return merged



def ordered_bounding_boxes(bboxes, x_thresh=200, y_thresh=100):
    if not bboxes:
        return []

    bboxes = np.array(bboxes)
    ordered_bboxes = []
    visited = np.zeros(len(bboxes), dtype=bool)

    def find_cluster(idx):
        cluster = [idx]
        x1, y1, x2, y2 = bboxes[idx]
        x_center, y_center = (x1 + x2) / 2, (y1 + y2) / 2

        for j in range(len(bboxes)):
            if j != idx and not visited[j]:
                x1_j, y1_j, x2_j, y2_j = bboxes[j]
                x_center_j, y_center_j = (x1_j + x2_j) / 2, (y1_j + y2_j) / 2

                dx = abs(x_center - x_center_j)
                dy = abs(y_center - y_center_j)

                if dx <= x_thresh and dy <= y_thresh:
                    visited[j] = True
                    cluster.extend(find_cluster(j))

        return cluster

    side_1 = []
    side_2 = []
    for i in range(len(bboxes)):
        if not visited[i]:
            visited[i] = True
            cluster = find_cluster(i)
            if len(side_1) == 0:
                for c in cluster:
                    x1, y1, x2, y2 = bboxes[c]
                    side_1.append((x1, y1, x2, y2))
            else:
                for c in cluster:
                    x1, y1, x2, y2 = bboxes[c]
                    side_2.append((x1, y1, x2, y2))

    if len(side_2) != 0 and side_1[0][0] < side_2[0][0]:
        ordered_bboxes.extend(side_1)
        ordered_bboxes.extend(side_2)

    else:
        ordered_bboxes.extend(side_2)
        ordered_bboxes.extend(side_1)

    return ordered_bboxes
```

```python
for x in os.walk(dataset_image_path):
    image_names = x[2]

print(image_names)
```

```python
predicted_text = []
ground_truth = []
for image_name in tqdm(image_names, desc="Images", unit="images"):

    bboxes = get_bboxes(image_name=image_name)
    merged_bboxes = merge_bounding_boxes(bboxes)
    ordered_bboxes = ordered_bounding_boxes(merged_bboxes)

    image_path = os.path.join(dataset_image_path, image_name)
    image = cv2.imread(image_path)

    label_name = image_name.split('.')[0] + ".txt"
    label_path = os.path.join(dataset_label_path, label_name)
    label = open(label_path, 'r', encoding="utf-8").read()
    label = label.split('\n')
    ground_truth.append(' '.join(label).lower())

    ocr_text = []

    for i, (x1, y1, x2, y2) in tqdm(enumerate(ordered_bboxes), unit="bboxes", desc="bboxes", total=len(ordered_bboxes)):
        cropped = image[y1:y2, x1:x2]

        cropped_pil = Image.fromarray(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))

        pixel_values = processor(images=cropped_pil, return_tensors="pt").pixel_values.to(device)

        with torch.no_grad():
            generated_ids = model.generate(pixel_values)

        recognized_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

        ocr_text.append(recognized_text)

    predicted_text.append(" ".join(ocr_text).lower())
```
